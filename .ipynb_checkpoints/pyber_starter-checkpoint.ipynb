Trends: 
1) As expected, Urban areas, which have the highest population density, had the most drivers, and the most number of rides.
2) Rural areas had the most expensive fares, however.  This makes sense considering that rural areas typically have longer distances between locations, which would increase the average fare.
3) An interesting trend, Suburban Drivers make up 16.5% of Pyber drivers, but they take on 26.3% of total rides.  Rural drivers make up only 2.6% of Pyber drivers, but take on 5.3% of total rides.   This suggests that rural and suburban drivers take on many more rides (per driver) than urban drivers do -- they have a higher rides:driver ratio.

```python

%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
cityData=pd.read_csv(city_data_to_load)
rideData=pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
pyberData = pd.merge(rideData, cityData, how='outer', on='city')

# Display the data table for preview
pyberData.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
# Build the scatter plots for each city types
Urban= pyberData.loc[pyberData["type"] == "Urban", :]
Rural= pyberData.loc[pyberData["type"] == "Rural", :]
Suburban= pyberData.loc[pyberData["type"] == "Suburban", :]

RuralGB= Rural.groupby(['city'])
R_yaxis=RuralGB['fare'].mean()
R_xaxis=RuralGB['ride_id'].count()
R_size=RuralGB['driver_count'].mean()*10

PyRural = plt.scatter(R_xaxis, R_yaxis, facecolors='gold', edgecolors="black", s=R_size, label="Rural", alpha=0.75)

UrbanGB= Urban.groupby(['city'])
U_yaxis=UrbanGB['fare'].mean()
U_xaxis=UrbanGB['ride_id'].count()
U_size=UrbanGB['driver_count'].mean()*10

PyUrban= plt.scatter(U_xaxis, U_yaxis, facecolors='coral', edgecolors="black", s=U_size, label="Urban", alpha=0.75)

SuburbanGB= Suburban.groupby(['city'])
S_yaxis=SuburbanGB['fare'].mean()
S_xaxis=SuburbanGB['ride_id'].count()
S_size=SuburbanGB['driver_count'].mean()*10

PySuburban= plt.scatter(S_xaxis, S_yaxis, facecolors='#87CEEB', edgecolors="black", s=S_size, label="Suburban", alpha=0.75)


# Incorporate the other graph properties
plt.grid()
plt.xlabel("Total Number of Rides (per city)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")

# Create a legend

lgnd= plt.legend(handles= [PyRural, PyUrban, PySuburban], labels=["Rural", "Urban", "Suburban"], loc="best", title="City Type")

for handle in lgnd.legendHandles:
    handle.set_sizes([40.0])

# Incorporate a text label regarding circle size
plt.text(43, 30, "Note: Circle size correlates with driver count per city", horizontalalignment='left', size='medium', color='black')


# Save Figure
plt.savefig("Pyber_Ride_Sharing_Data")
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
GroupbyType = pyberData.groupby("type")
fares = [element for element in GroupbyType.sum()["fare"]]

# Build Pie Chart
CityType = ['Rural', 'Suburban', 'Urban']
Colors = ['gold',"#87CEEB", 'coral']
explode= [0, 0, .1]


plt.pie(fares, labels=CityType, colors=Colors, explode=explode, autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("auto")
plt.title("% of Fare Revenue by City Type")



# Save Figure
plt.savefig("Fare_Pie_Chart")
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
rides = [element for element in GroupbyType.count()["ride_id"]]

# Build Pie Chart
plt.pie(rides, explode=explode, colors=Colors, labels=CityType, autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("auto")
plt.title("% of Total Rides by City Type")



# Save Figure
plt.savefig("Total_Rides_Pie_Chart")
plt.show()
```
